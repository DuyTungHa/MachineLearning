```python
# Import relevant packages
import numpy as np
import pandas as pd
import tensorflow.compat.v1 as tf
from tensorflow.keras.layers import Dense

# Disable tensorflow v2 and reset default graph
tf.disable_v2_behavior()
tf.reset_default_graph()
```

```python
# Load data into pandas dataframe
col_names = ["class", "lepton_1_pT", "lepton_1_eta", "lepton_1_phi", "lepton_2_pT", "lepton_2_eta", "lepton_2_phi", "missing_energy_magnitude", "missing_energy_phi", "MET_rel", "axial_MET", "M_R", "M_TR_2", "R", "MT2", "S_R", "M_DELTA_R", "dPhi_r_b", "cos_thera_r1"]
SUSY = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/SUSY.csv', names=col_names, sep=',', header=None)
```

```python
# Split data into training set and test set
train_SUSY = SUSY.iloc[:4500000, :]
test_SUSY = SUSY.iloc[4500000:, :]
train_X = train_SUSY.iloc[:, 1:].to_numpy()
train_Y = train_SUSY.iloc[:, :1].to_numpy()
test_X = test_SUSY.iloc[:, 1:].to_numpy()
test_Y = test_SUSY.iloc[:, :1].to_numpy()
```

```python
# Set model and training parameters
n_epochs = 12
learning_rate = 0.5
n_neurons = 1
batch_size = 300000

# Create and compile a training model
y_hat_K = tf.keras.Sequential([Dense(n_neurons, activation="sigmoid", kernel_regularizer=tf.keras.regularizers.l2(0.01))])
y_hat_K.compile(optimizer=tf.train.GradientDescentOptimizer(learning_rate), loss="binary_crossentropy", metrics=['accuracy'])

# Train the model and evaluate against the test set
y_hat_K.fit(train_X, train_Y, epochs=n_epochs, batch_size=batch_size)
y_hat_K.evaluate(test_X, test_Y)
```
